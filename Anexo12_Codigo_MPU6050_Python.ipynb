# figuras del acelerómetro

```python
# @title Figura version 1
# Datos de las señales
y1 = df.ax
y2 = df.ay
y3 = df.az
x = np.linspace(1, len(y1), len(y1))

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(15,5))

# Graficar las señales
ax.plot(x, y1, color='royalblue', label='ax')
ax.plot(x, y2, color='orange', label='ay')
ax.plot(x, y3, color='gray', label='az')

# Configurar título y etiquetas de los ejes
ax.set_title('MPU 6050 - Acelerometro')
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Aceleración [m.s^-2]')

# Agregar cuadrícula en el eje Y
ax.grid(axis='y')

# Mostrar la leyenda
ax.legend()

ax.scatter(x, y1, color='royalblue', label='Señal 1')
ax.scatter(x, y2, color='orange', label='Señal 2')
ax.scatter(x, y3, color='gray', label='Señal 3')

# Mostrar la figura
plt.show()
```

```python
# @title Figura version 2
# Datos de las señales
y1 = df.ax
y2 = df.ay
y3 = df.az
x = np.linspace(1, len(y1), len(y1))

# Crear las trazas de las señales
trace1 = go.Scatter(x=x, y=y1, mode='lines', name='Señal 1', line=dict(color='royalblue'))
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='Señal 2', line=dict(color='orange'))
trace3 = go.Scatter(x=x, y=y3, mode='lines', name='Señal 3', line=dict(color='gray'))
trace1_= go.Scatter(x=x, y=y1, mode='markers', name='Señal 1', marker=dict(color='royalblue'), showlegend=False)
trace2_= go.Scatter(x=x, y=y2, mode='markers', name='Señal 2', marker=dict(color='orange'), showlegend=False)
trace3_= go.Scatter(x=x, y=y3, mode='markers', name='Señal 3', marker=dict(color='gray'), showlegend=False)
# Crear el layout de la figura
layout = go.Layout(
    title='Visualización de las señales',
    xaxis=dict(title='Eje X'),
    yaxis=dict(title='Eje Y', showgrid=True)
)

# Crear la figura y agregar las trazas
fig = go.Figure(data=[trace1, trace2, trace3, trace1_, trace2_, trace3_], layout=layout)

# Mostrar la figura
# fig.show()
fig.show()
```
